```python
import pandas as pd
import numpy as np
```

```python
churn = pd.read_csv("D:\DATA SCIENCE ACADEMY\Pre Test\churn.csv", encoding='latin1')
churn.head()
```

```python
churn.describe()
```

```python
# Insight(1) Info mengenai jumlah pelanggan yang churn dan tidak churn
churn['class'].value_counts()
```

```python
# Insight(2) Info mengenai state (negara bagian) mana yg mempunyai populasi churn dan stay terbanyak
data1 = churn[['state', 'class']]
xf = data1.sort_values('class',ascending=False)
state1 = xf['class'] == 1
state0 = xf['class'] == 0
print("State paling banyak churn")
print(xf[state1].mode())
print("State paling banyak stay")
print(xf[state0].mode())
```

```python
# Insight(3) Info mengenai profiling pelanggan churn dan stay dari parameter International Plan
data2 = churn[['international_plan', 'total_intl_minutes', 'total_intl_calls', 'total_intl_charge', 'class']]
yf = data2.sort_values('class',ascending=False)
state11 = yf['class'] == 1
state00 = yf['class'] == 0
print("analisa mode parameter international plan status churn")
print(yf[state11].mode())
print("analisa mean parameter international plan status churn")
print(yf[state11].mean())
print("analisa std parameter international plan status churn")
print(yf[state11].std())
print("analisa mode parameter international plan status stay")
print(yf[state00].mode())
print("analisa mean parameter international plan status stay")
print(yf[state00].mean())
print("analisa std parameter international plan status stay")
print(yf[state00].std())
```

```python
# Insight(4) Info mengenai profiling pelanggan churn dan stay dari parameter paket Day
data3 = churn[['total_day_minutes', 'total_day_calls', 'total_day_charge', 'class']]
zf = data3.sort_values('class',ascending=False)
state111 = zf['class'] == 1
state000 = zf['class'] == 0
print("analisa mode parameter day usage status churn")
print(zf[state111].mode())
print("analisa mean parameter day usage status churn")
print(zf[state111].mean())
print("analisa std parameter day usage status churn")
print(zf[state111].std())
print("analisa mode parameter day usage status stay")
print(zf[state000].mode())
print("analisa mean parameter day usage status stay")
print(zf[state000].mean())
print("analisa std parameter day usage status stay")
print(zf[state000].std())
```

```python
# Insight(5) Info mengenai profiling pelanggan churn dan stay dari parameter voice_mail & cust_service
data4 = churn[['voice_mail_plan', 'number_customer_service_calls', 'class']]
bf = data4.sort_values('class',ascending=False)
state1111 = zf['class'] == 1
state0000 = zf['class'] == 0
print("analisa mode parameter voice_mail & cust_service status churn")
print(bf[state1111].mode())
print("analisa mean parameter voice_mail & cust_service status churn")
print(bf[state1111].mean())
print("analisa std parameter voice_mail & cust_service status churn")
print(bf[state1111].std())
print("analisa mode parameter voice_mail & cust_service status stay")
print(bf[state0000].mode())
print("analisa mean parameter voice_mail & cust_service status stay")
print(bf[state0000].mean())
print("analisa std parameter voice_mail & cust_service status stay")
print(bf[state0000].std())
```
